# Conditional Statements
Conditional Statements are statements in Python that provide a choice for the control flow based on a condition. It means that the control flow of the Python program will be decided based on the outcome of the condition.
Types of conditional statements:


*If--
If the simple code of block is to be performed if the condition holds then the if statement is used


*If-Else--In a conditional if Statement the additional block of code is merged as an else statement which is performed when if condition is false.

*If-Elif--As soon as one of the conditions controlling the if is true, the statement associated with that if is executed, and the rest of the ladder is bypassed. If none of the conditions is true, then the final else statement will be executed.







````python
# if statement example
if 10 > 5:
    print("10 greater than 5")

print("Program ended")
````

````python
# if..else statement example
x = 3
if x == 4:
    print("Yes")
else:
    print("No")
````

````python
# if..else chain statement
letter = "A"

if letter == "B":
    print("letter is B")

else:

    if letter == "C":
        print("letter is C")

    else:

        if letter == "A":
            print("letter is A")

        else:
            print("letter isn't A, B and C")
````

````python
# if-elif statement example
letter = "A"

if letter == "B":
    print("letter is B")

elif letter == "C":
    print("letter is C")

elif letter == "A":
    print("letter is A")

else:
    print("letter isn't A, B or C")
````

**break Statement**

With the break statement we can stop the loop before it has looped through all the items.

**continue Statement**

With the continue statement we can stop the current iteration of the loop, and continue with the next:



````python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    break
  print(x)
````

Accessing List Elements

Elements in a list can be accessed using indexing. Python indexes start at 0, so a[0] will access the first element, while negative indexing allows us to access elements from the end of the list

````python
a = [10, 20, 30, 40, 50]

print(a[0])


print(a[-1])
````

We can add elements to a list using the following methods:

**append(): Adds an element at the end of the list.**

syntax--(your list).append(value)

**extend(): Adds multiple elements to the end of the list.**

syntax--(your list).extend(another list)

**insert(): Adds an element at a specific position.**

syntax--(your list).insert(index,value)

````python

a = []
# Adding 10 to end of list
a.append(10)
print("After append(10):", a)

# Inserting 5 at index 0
a.insert(0, 5)
print("After insert(0, 5):", a)

# Adding multiple elements  [15, 20, 25] at the end
a.extend([15, 20, 25])
print("After extend([15, 20, 25]):", a)
````

We can remove elements from a list using:

**remove(): Removes the first occurrence of an element.**

syntax--(your list).remove(value)

**pop(): Removes the element at a specific index or the last element if no index is specified.**

syntax--(your list).pop()

**del statement: Deletes an element at a specified index.**

syntax--del (your list)[index]

````python
a = [10, 20, 30, 40, 50]

# Removes the first occurrence of 30
a.remove(30)
print("After remove(30):", a)

# Removes the element at index 1 (20)
popped_val = a.pop(1)
print("Popped element:", popped_val)
print("After pop(1):", a)

# Deletes the first element (10)
del a[0]
print("After del a[0]:", a)
````

# Slicing
Slicing a Python tuple means dividing a tuple into small tuples using the indexing method.It is helpful in obtaining desired part of the list,tuple,string etc



```
(your list/string)[startindex:endindex:jump]
```
where:

startindex  is the starting index

endindex  is the end index-1

jump  is the jump between indexes



````python
# code to test slicing
t = (0 ,1, 2, 3)

print(t[1:])
print(t[::-1])
print(t[2:4])
print(t[1:3])
````

````python
b = "Hello, World!"
print(b[-5:-2])
````

**Negative Slicing**


```
(your dataset)[:-2]
```
is same as


```
(your dataset)[:len(your dataset)-2]
```





````python
s = "abcdefghijklmno"

# Characters from index -4 to the end
print(s[-4:])

# Characters from the start up to index -3 (excluding -3)
print(s[:-3])

# Characters from index -5 to -2 (excluding -2)
print(s[-5:-2])

# Get every 2nd elements from index -8 to -1 (excluding index -1)
print(s[-8:-1:2])
````
